```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('C:/Users/hargun kaur/OneDrive/Desktop/gender_submission.csv')

# Perform data cleaning

# Drop irrelevant columns
df = df.drop(['PassengerId'], axis=1)

# Perform EDA

# Explore the variable 'Survived'
survived_counts = df['Survived'].value_counts()
print('Survived Counts:')
print(survived_counts)

# Visualize the variable 'Survived'
sns.countplot(x='Survived', data=df)
plt.title('Survived Counts')
plt.show()

# Explore the relationship between 'PassengerId' and 'Survived'
sns.scatterplot(x='PassengerId', y='Survived', data=df)
plt.title('Survived by PassengerId')
plt.show()
```
